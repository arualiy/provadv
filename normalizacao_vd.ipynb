```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from kaggle.api.kaggle_api_extended import KaggleApi

# Autenticar no Kaggle
os.environ['KAGGLE_CONFIG_DIR'] = r'C:\Users\valde\Downloads\prova_dv'
api = KaggleApi()
api.authenticate()

# Baixar o dataset
dataset = 'umerhaddii/paris-olympic-medalists-2024'
api.dataset_download_files(dataset, path='datasets/', unzip=True)

# Carregar o dataset
file_path = r'C:\Users\valde\Downloads\prova_dv\datasets\2024_medalists_all.csv' 
data = pd.read_csv(file_path)

# Verifique se o DataFrame foi carregado corretamente
print(data.head())
```

```python
# País
countries_df = data[['country_medal_wikidata_id', 'country_medal']].drop_duplicates()
countries_df = countries_df.rename(columns={
    'country_medal_wikidata_id': 'country_id',
    'country_medal': 'country_name'
})
countries_df
```

```python
# Delegação
delegations_df = data[['delegation_wikidata_id', 'delegation_name']].drop_duplicates()
delegations_df = delegations_df.rename(columns={'delegation_wikidata_id': 'delegation_id'})
delegations_df
```

```python
# Esportes
sports_df = data[['sport_wikidata_id', 'sport', 'event_wikidata_id']].drop_duplicates()
sports_df = sports_df.rename(columns={
    'sport_wikidata_id': 'sport_id',
    'event_wikidata_id': 'event_id',
})
sports_df
```

```python
# Medalhista
medalists_df = data[['medalist_wikidata_id','country_medal_wikidata_id', 'medalist_name', 'date_of_birth', 'sex_or_gender', 'place_of_birth_wikidata_id', 'delegation_wikidata_id','event_wikidata_id', 'sport_wikidata_id', 'medal']].drop_duplicates()
medalists_df = medalists_df.rename(columns={
    'medalist_wikidata_id': 'medalist_id',
    'place_of_birth_wikidata_id': 'birthplace_id',
    'sex_or_gender': 'gender',
    'delegation_wikidata_id': 'delegation_id',
    'event_wikidata_id': 'event_id',
    'sport_wikidata_id': 'sport_id',
    'country_medal_wikidata_id': 'country_id'
})
medalists_df
```

```python
# Local de Nascimento
birthplaces_df = data[['place_of_birth_wikidata_id', 'place_of_birth', 'lat', 'lon']].drop_duplicates()
birthplaces_df = birthplaces_df.rename(columns={'place_of_birth_wikidata_id': 'birthplace_id'})
birthplaces_df
```

```python
medals_df = data[['medal']].drop_duplicates()
medals_df
```

```python
medals_df.to_excel('medals.xlsx', index=False)
birthplaces_df.to_excel('medals.xlsx', index=False)
medalists_df.to_excel('medalists.xlsx', index=False)
sports_df.to_excel('sports.xlsx', index=False)
delegations_df.to_excel('delegations.xlsx', index=False)
countries_df.to_excel('countries.xlsx', index=False)
```
